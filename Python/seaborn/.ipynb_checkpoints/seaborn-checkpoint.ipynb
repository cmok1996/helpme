```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
```

```python
crash_df.head()
```

## Distribution Plots

```python
sns.displot(crash_df["not_distracted"], kde=False, bins=25)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df, hue='sex')
```

```python
sns.rugplot(tips_df['tip'])
```

## Palettes

```python
tips_df = sns.load_dataset('tips')

sns.set_style('ticks') #darkgrid, whitegrid, dark, ticks

plt.figure(figsize = (8,6))

sns.set_context(context = 'paper', font_scale = 1.4)
sns.stripplot(x='day', y='total_bill', data=tips_df, palette = 'afmhot', hue='sex') #search matplotlib colourmaps documentation

sns.despine(left=False, bottom=True) #hide grid lines

ax = plt.gca()
ax.set_ylabel("Total Bill")
ax.set_xlabel("Days")

plt.legend(loc=1)
```

## Categorical data

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator = np.median)
```

```python
sns.countplot(x='sex', data=tips_df)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.legend(loc=1)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

```python
plt.figure(figsize = (10,6))
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data = tips_df, color='white')
```

## Matrix plots

```python
plt.figure(figsize = (8,6))
sns.set_context('paper', font_scale = 1.4)

crash_mx = crash_df.corr()
sns.heatmap(crash_mx, annot=True, cmap = 'viridis')
```

```python
flights_df = sns.load_dataset('flights')
flights_df
```

```python
flights = flights_df.pivot_table(index = 'year', columns='month', values = 'passengers')
sns.heatmap(flights, cmap="Blues", linecolor = 'white', linewidth = 2)
```

## Cluster

```python
sns.clustermap(flights, cmap='Blues', standard_scale = True)
```

## Pair grids

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species', x_vars = ['petal_length', 'sepal_width'],
                     y_vars = ['petal_length', 'sepal_width']) 
#iris_g.map_offdiag(plt.scatter)
iris_g.map_diag(sns.histplot)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
iris_g.add_legend()
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height = 4, aspect=1.3, col_order = ['Dinner', 'Lunch'], 
                       hue_order = ['No', 'Yes'],
                       hue_kws = dict(marker = ['^', 'v']))

#tips_fg.map(plt.hist, 'total_bill', bins=8)
kws = dict(s=50, linewidth=.5)
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
tips_fg.add_legend()
```

## Regression plots

```python
plt.figure(figsize = (8,6))
sns.set_context('paper', font_scale = 1.4)
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips_df, markers = ['o', '^'],
          scatter_kws = {'s':50, 'linewidth':0.5})
```

```python
sns.lmplot(x='total_bill', y='tip', col='sex', row='time', data=tips_df)
```
